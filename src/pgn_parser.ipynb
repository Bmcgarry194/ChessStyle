```python
def list_of_games_in_file(pgn_file):
    with open(pgn_file) as pgn:
        game_list = []
        while chess.pgn.read_game(pgn):
            game_list.append(chess.pgn.read_game(pgn))
        return game_list[:-1]
       

def headers_to_dict(game):
    result = {}
    labels = ['Event', 'Site', 'Date', 'Round', 'White', 'Black', 'Result']
    for label in labels:
        result[label] = game.headers[label]
    result['pgn'] = [str(move) for move in game.main_line()]
    return result

def pgn_file_to_df(pgn_file):
    games = list_of_games_in_file(pgn_file)
    games_list = []
    for game in games:
        games_list.append(headers_to_dict(game))
    return pd.DataFrame(games_list)
```

```python
def load_chess_stats_json(file):
    with open(file, 'r') as f:
        data = json.loads(f.read())
    return data

def castling_stats(data):
    castling = {}
    castling['k_black'] = data['Castling']['Black']['Kingside']
    castling['q_black'] = data['Castling']['Black']['Queenside']
    
    castling['k_white'] = data['Castling']['White']['Kingside']
    castling['q_white'] = data['Castling']['White']['Queenside']
    
    castling['opposite_side'] = data['Castling']['Side']['Opposite']
    castling['same_side'] = data['Castling']['Side']['Same']
    
    return castling

def total_games(data):
    data['totalGames']
    
```
